### ii. Extract the time-domain features minimum, maximum, mean, median, standard deviation, first quartile, and third quartile for all of the 6 time series in each instance.

```python
# Detect missing values(NaN)
for f in files_all:
    df1 = pd.read_csv(f, skiprows=[0,1,2,3,4],sep='\s|,',engine='python',
                  names=['# Columns: time','avg_rss12','var_rss12','avg_rss13','var_rss13','avg_rss23','var_rss23'] )
    df2 = df1.drop(df1.columns[0],axis=1)
    if df2.isnull().values.any() == True:
        print('There are missing values.')
```

```python
n = 0
list_new = []

try:
    for f in files_all:
        n = n + 1
    #df1 = pd.read_csv(f, skiprows=[0,1,2,3])
        df1 = pd.read_csv(f, skiprows=[0,1,2,3,4],sep='\s|,',engine='python',
                  names=['# Columns: time','avg_rss12','var_rss12','avg_rss13','var_rss13','avg_rss23','var_rss23'] )
        df2 = df1.drop(df1.columns[0],axis=1)
        l_min= list(df2.min())
        l_max=list(df2.max())
        l_mean=list(df2.mean(numeric_only=False))
        l_median=list(df2.median(numeric_only=False))
        l_std=list(df2.std(numeric_only=False))
        l_1st=list(df2.quantile(q=0.25))
        l_3rd=list(df2.quantile(q=0.75))
        dic_f = {}
        dic_f['Instance'] = [n]
        for i in range(1,7):
            dic_f[f'min{i}'] = [l_min[i-1]]
            dic_f[f'max{i}'] = [l_max[i-1]]
            dic_f[f'mean{i}'] = [l_mean[i-1]]
            dic_f[f'median{i}'] = [l_median[i-1]]
            dic_f[f'std{i}'] = [l_std[i-1]]
            dic_f[f'1st quart{i}'] = [l_1st[i-1]]
            dic_f[f'3rd quart{i}'] = [l_3rd[i-1]]
        list_new.append(pd.DataFrame(dic_f))
        
except:
    print(f)

else:
    df_n = pd.concat(list_new, ignore_index=True)
```

### iii. Estimate the standard deviation of each of the time-domain features you extracted from the data. Then, use Python’s bootstrapped or any other method to build a 90% bootsrap confidence interval for the standard deviation of each feature.

```python
# the standard deviation of each of the time-domain features
df_std = pd.DataFrame(df_n.std()[1:],columns=["std"])
df_std
```

```python
# Use Python’s bootstrapped to build a 90% confidence interval for the standard deviation of each feature.
from scipy.stats import bootstrap
import numpy as np

list_ci = []
for i in range(1,43):
    data = list(df_n.iloc[:,i])
    data = (data,)
    bootstrap_result = bootstrap(data, np.std, confidence_level=0.9, random_state=1, method='percentile')
    list_ci.append(bootstrap_result.confidence_interval)
for i in range(1,43):
    print(df_n.columns.values[i],':',list_ci[i-1])
```
